```python
import plotnine as pn
from plotnine import ggplot, geom_line, geom_point, aes
import polars as pl
```

```python
toy_data = pl.DataFrame({"A" : (2,1,4,8),
                         "B" : (4,1,15,80),
                         "C" : (1,2,3,4),
                         "D" : ('far','far','near','near')})
toy_data
```

```python
# Plot 1 (Top left)

(
    ggplot(toy_data)
    + geom_point(
        aes(x="A", y="B", color="D", shape="D"),
        size=3
    )
    + geom_line(
        aes(x="A", y="B"),
        color="black",
        size=1.2
    )
)
```

```python
# Plot 2 (Top right)

(
    ggplot(toy_data)
    + geom_line(
        aes(x="A", y="B", group="D", color="D"),
        size=1.2
    )
    + geom_point(
        aes(x="A", y="B", shape="D"),
        color="black",
        size=3
    )
)
```

```python
#Plot 3 (Bottom left)

(
    ggplot(toy_data)
    + geom_line(
        aes(x="A", y="B"),
        color="black",
        size=1.2
    )
    + geom_point(
        aes(x="A", y="B", color="D", shape="D"),
        size=3
    )
)
```

```python
# Plot 4 (Bottom right)

(
    ggplot(toy_data)
    + geom_line(
        aes(x="A", y="B", group="D", color="D", linetype="D"),
        size=1.2
    )
    + geom_point(
        aes(x="A", y="B"),
        color="black",
        size=3
    )
)
```

```python
mexico_hc_spending = pl.read_csv("https://github.com/yardsale8/DSCI_210_R_notebooks/raw/main/data/Mexico_Health_Spending.csv")
mexico_hc_spending = (mexico_hc_spending
                      .rename({'Spending (in billions, 2017 Purchasing Power Parity Dollars PPP $)':'Spending'})
                     )
```

```python
from plotnine import ggplot, geom_bar, aes, facet_grid

(
    ggplot(data=mexico_hc_spending)
    + geom_bar(
        aes(
            x="Year",
            y="Spending",
            fill = "Type"
        ),
        stat="identity"
    )
    + facet_grid(rows = "Type") 
)
```
