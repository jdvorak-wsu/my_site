```python
import polars as pl
import plotnine as pn
```

```python
pn.options.figure_size = (6,4)
```

```python
mexico_hc_spending = pl.read_csv("https://github.com/yardsale8/DSCI_210_R_notebooks/raw/main/data/Mexico_Health_Spending.csv")
mexico_hc_spending.head()
```

```python
mexico_hc_spending.columns
```

```python
mexico_hc_spending = (mexico_hc_spending
                      .rename({'Spending (in billions, 2017 Purchasing Power Parity Dollars PPP $)':'Spending'})
                     )
                     
```

```python
mexico_hc_spending.head()
```

```python
(
    pn.ggplot(data=mexico_hc_spending)
    + pn.geom_bar(
        pn.aes(
            x="Year",
            y="Spending",
            fill="Type"
        ),
        stat="identity",
        position="stack"
    )
)

                     
                     
```

```python
from plotnine import ggplot, geom_bar, aes

(
    ggplot(data=mexico_hc_spending)
    + geom_bar(
        aes(
            x="Year",
            y="Spending",
            fill="Type"
        ),
        stat="identity",
        position="stack"
    )
)

               
        
```

```python
##QUESTION ONE: TRANSLATE R CODE INTO PYTHON

(
    pn.ggplot(data=mexico_hc_spending)
    + pn.geom_line(
        pn.aes(
            x="Year",
            y="Spending",
            color="Type"
        )
    )
    + pn.geom_point(
        pn.aes(
            x="Year",
            y="Spending",
            color="Type"
        )
    )
)
```

```python
## QUESTION TWO: from plotnine import ......

from plotnine import ggplot, geom_line, geom_point, aes

(
    ggplot(data=mexico_hc_spending)
    + geom_line(
        aes(
            x="Year",
            y="Spending",
            color="Type"
        )
    )
    + geom_point(
        aes(
            x="Year",
            y="Spending",
            color="Type"
        )
    )
)
```

```python
births = pl.read_csv('https://github.com/yardsale8/DSCI_210_R_notebooks/raw/main/data/NC_Birth.csv', null_values=['.'])
births.head
```

```python
from plotnine import ggplot, geom_point, aes, labs

(ggplot(births)
    + geom_point(
        aes(
            x="Mother Age",
            y="Gestational  Age",
            color="Mother Smoker"
           )
      )
    + labs(
        x="Mother Age",
        y="Gestational Age",
        color="Mother Smoker"
      )
)
```
